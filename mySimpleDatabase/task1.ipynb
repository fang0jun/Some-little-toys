```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
influence_data=pd.read_csv('influence_data.csv')
influence_data.head()
```

#### influencer_ id: 影响者的唯一标识号。 ( 数字字符串)
#### influencer_ name: 追随者或行业专家给出的影响艺术家的名称。(字符串)
#### influencer_main_genre:最能描述这位有影响力的艺术家创作的大部分音乐的流派。(如果有) (字符串)
#### influencer_ active_ start: 这位有影响力的艺术家开始音乐生涯的年代。(整数)
#### follower_ id:追随者的艺术家的唯一标识号。 (数字字符串)
#### follower_ name:追随有影响力艺术家的艺术家名称。(字符串)
#### follower_ main_ genre: 最能描述以下艺术家创作的大部分音乐的流派。(如果有)(字符串)
#### follower_ _active_ start: 追随艺术家开始音乐生涯的年代。(整数)

```python
influence_data.shape
```

```python
data=influence_data[['influencer_name','follower_name']]
data['weight']=1
data.head()
```

```python
from collections import Counter
sorted(Counter(data['influencer_name']).items(),key=lambda x:x[1],reverse=True)
```

```python
num=300
node1=data['influencer_name'].values.tolist()[:num]
node2=data['follower_name'].values.tolist()[:num]
nodeweight=data['weight'].values.tolist()[:num]
G=nx.Graph()
for i in range(len(node1)):
    G.add_edge(node1[i],node2[i], weight=nodeweight[i])
    
nx.draw(G, with_labels=True)
plt.savefig('3.jpg')
```

```python
data.shape[0]
```

```python
num=300
all_node=list(set(data['influencer_name'].values.tolist()[:num]  +data['follower_name'].values.tolist()[:num]))
all_list=[]
node1=data['influencer_name'].values.tolist()[:num]
node2=data['follower_name'].values.tolist()[:num]
nodeweight=data['weight'].values.tolist()[:num]
for i in range(len(node1)):
    all_list.append((node1[i],node2[i],nodeweight[i]))
```

```python
num=300
node1=data['influencer_name'].values.tolist()[:num]
node2=data['follower_name'].values.tolist()[:num]
nodeweight=data['weight'].values.tolist()[:num]
G=nx.Graph()
for i in range(len(node1)):
    G.add_edge(node1[i],node2[i], weight=nodeweight[i])
    
nx.draw(G, with_labels=True)
plt.savefig('3.jpg')
```

```python
sorted(G.degree(),key=lambda x: x[1],reverse=True)
```

```python
%matplotlib notebook

#plt.figure(figsize=(100,100))
#解决中文显示问题
plt.rcParams['font.sans-serif'] = ['KaiTi'] # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
net_grid = nx.Graph()    # 新建一个无向图 net_grid

# 网络图中的所有的节点，以列表的形式存储
list_net_nodes = all_node
 
# 网络图中的所有的边，以列表的形式存储，由于是无向图，可以自动忽略掉重读的边
list_net_edges =all_list
 
# 在网络图中添加节点，直接从节点列表中读取所有的节点
net_grid.add_nodes_from(list_net_nodes)
 
# 在网络中添加边，直接从边列表中读取所有的边
net_grid.add_edges_from(list_net_edges)
 
# 画网络图
nx.draw(net_grid, with_labels=True)
 
# 显示
plt.show()
```
